```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os, datetime, warnings
print('Last updated on {}'.format(datetime.datetime.now().ctime()))
```

```python
import sns_settings
sns.set_context('notebook')
pd.options.display.max_columns = 50
warnings.filterwarnings('ignore')
```

## DIC predictions

### Model including all available features

```python
df_test_dic1 = pd.read_csv('results/bottle_data_test_dic_all_vars.csv', index_col=0)

df_work = pd.melt(df_test_dic1, value_vars=df_test_dic1.columns.tolist()[:8], id_vars=['DIC residuals'])
g = sns.relplot(x='value', y='DIC residuals', data=df_work, col='variable', col_wrap=4, facet_kws={'sharex':False})
_ = g.set_titles(template='{col_name}', row_template='{col_name}')
```

### Model excluding $O_2$

```python
df_test_dic2 = pd.read_csv('results/bottle_data_test_dic_noO2.csv', index_col=0)

df_work = pd.melt(df_test_dic2, value_vars=df_test_dic2.columns.tolist()[:7], id_vars=['DIC residuals'])
g = sns.relplot(x='value', y='DIC residuals', data=df_work, col='variable', col_wrap=4, facet_kws={'sharex':False})
_ = g.set_titles(template='{col_name}', row_template='{col_name}')
```

### Model excluding $O_2$ and satellite data

```python
df_test_dic3 = pd.read_csv('results/bottle_data_test_dic_nosat.csv', index_col=0)

df_work = pd.melt(df_test_dic3, value_vars=df_test_dic3.columns.tolist()[:4], id_vars=['DIC residuals'])
g = sns.relplot(x='value', y='DIC residuals', data=df_work, col='variable', col_wrap=4, facet_kws={'sharex':False})
_ = g.set_titles(template='{col_name}', row_template='{col_name}')
```
